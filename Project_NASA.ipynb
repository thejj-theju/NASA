```python
# using loop to get 10,000 data
from datetime import datetime
asteroids_data = []
target = 10000
API_KEY = "3damZYCyNWx8v5j7ZNDb8TdIYhuv1QFP4qt0UkZR"

import requests
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2025-05-01&end_date=2025-05-08&api_key={API_KEY}"


while len(asteroids_data)< target:
    response=requests.get(url)
    data=response.json()
    neo=data['near_earth_objects']

    for date,ast_details in neo.items():   # for key,value in dict.items():
        for ast in ast_details:   #for var in value: --> loop each item in value
            close_approach =ast['close_approach_data'][0]['close_approach_date']
                # Convert string to datetime object
            close_approach_date = datetime.strptime(close_approach, "%Y-%m-%d").strftime("%Y-%m-%d")
            asteroids_data.append(dict(
                id=int(ast['id']),
                name=ast['name'],
                neo_id=int(ast['neo_reference_id']),
                absolute_magnitude_h=ast['absolute_magnitude_h'],
                estimated_dia_min=ast['estimated_diameter']['kilometers']['estimated_diameter_min'],
                estimated_dia_max=ast['estimated_diameter']['kilometers']['estimated_diameter_max'],
                hazard=ast['is_potentially_hazardous_asteroid'],
                close_approach_date=close_approach_date,
                relative_velocity_kmph=float(ast['close_approach_data'][0]['relative_velocity']['kilometers_per_hour']),
                astronomical=float(ast['close_approach_data'][0]['miss_distance']['astronomical']),
                miss_distance_km=float(ast['close_approach_data'][0]['miss_distance']['kilometers']),
                miss_distance_lunar=float(ast['close_approach_data'][0]['miss_distance']['lunar']),
                orbiting_body=ast['close_approach_data'][0]['orbiting_body']
                
            ))
           
            if len(asteroids_data)==target:
                break
        if len(asteroids_data)==target:
            break
    url=data['links']['next']
```
